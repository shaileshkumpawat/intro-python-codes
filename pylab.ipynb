```python
import pylab as plt
```

```python
sample = []
linear = []
quadratic = []
cubic = []
exp = []

for i in range (-30,30):
    sample.append(i)
    linear.append(i)
    quadratic.append(i**2)
    cubic.append(i**3)
    exp.append(2**i)
```

```python
plt.plot(sample, linear)
plt.plot(sample, quadratic)
```

```python
plt.plot(sample, quadratic)
```

```python
plt.plot(sample, cubic)
```

```python
plt.plot(sample, exp)
```
